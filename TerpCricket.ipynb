# **Entity Definitions for TerpCricket**

## **1) Customers**  
Customers are cricket enthusiasts from all over **D.C. and Maryland** who visit **College Park** to find quality cricket equipment that is otherwise unavailable in most other stores. Given the growing **South Asian and international student population**, there is a strong demand for cricket gear, and TerpCricket is uniquely positioned to serve this niche market.

# **Relationships Between Entities in TerpCricket**

## **1) Customers (1:M) Orders**
- **One customer** can place **multiple orders**.
- **Each order** belongs to only **one customer**.
- This relationship ensures that we can track all purchases made by a specific customer.

**Foreign Key:** `Orders.customer_id ‚Üí Customers.customer_id`

---

## **2) Orders (1:M) Order_Items**
- **One order** can contain **multiple products**.
- **Each entry in Order_Items** represents a **product purchased in an order**.
- This allows us to track the **quantity and subtotal** for each product within an order.

**Foreign Key:** `Order_Items.order_id ‚Üí Orders.order_id`

---

## **3) Products (1:M) Order_Items**
- **One product** can be purchased in **multiple orders**.
- **Each entry in Order_Items** links a **product to a specific order**.
- This ensures that we can analyze product sales performance.

**Foreign Key:** `Order_Items.product_id ‚Üí Products.product_id`

---

## **4) Orders (M:1) Employees**
- **Each order** is processed by **one employee**.
- **One employee** can handle **multiple orders**.
- This ensures that employee productivity and order management can be tracked.

**Foreign Key:** `Orders.employee_id ‚Üí Employees.employee_id`

---

## **5) Products (M:1) Suppliers**
- **One supplier** can provide **multiple products**.
- **Each product** is sourced from **one supplier**.
- This ensures we can track where each product is coming from.

**Foreign Key:** `Products.supplier_id ‚Üí Suppliers.supplier_id`

---

## **Summary of Relationships**
| **Relationship**  | **Cardinality** | **Description** |
|------------------|---------------|----------------|
| **Customers ‚Üî Orders** | 1:M | One customer can place multiple orders. |
| **Orders ‚Üî Order_Items** | 1:M | One order can contain multiple products. |
| **Products ‚Üî Order_Items** | 1:M | One product can appear in multiple orders. |
| **Orders ‚Üî Employees** | M:1 | One employee processes multiple orders. |
| **Products ‚Üî Suppliers** | M:1 | One supplier provides multiple products. |

#**Normalization of TerpCricket's Relational Database Schema**

## **First Normal Form**


*   All columns contain atomic values [no lists or mulitple values in a single field ].
*   Each row has a unique identifier [ Primary Key ].
*   No duplicate Rows.

Since all our tables have a primary key, and all of our attributes contain atomic values and no duplicate rows, it satisfies First Normal Form

## **Second Normal Form**

*   Must be in First Normal Form.
*   Must not contain partial dependencies [ i.e. all non-key columns must depend on the entire primary key, and not just a part of it ].

Since our tables with single-column primary keys, every column fully depends no the Primary Key, and no column is partially depended on just one part of a composite key[ even Order_Items table which has order_id and product_id as a composite primary key, all the attributes fully depend on both of these keys ], it satisfies the Second normal form.

## **Third Normal Form**

*   Must satisfy Second Normal Form.
*   Must have non transitive dependencies [ A non key column must depend only on primary key and not any other non key column itself ]

Since our tables follow these and have no unnecessary dependencies either, our whole database follows **Third Normal Form**

```python
# Define table creation queries
table_queries = {
    "Customers": """
    CREATE TABLE IF NOT EXISTS Customers (
        customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
        first_name TEXT NOT NULL,
        last_name TEXT NOT NULL,
        email TEXT UNIQUE NOT NULL,
        phone_number TEXT NOT NULL,
        join_date DATE NOT NULL
    );
    """,

    "Products": """
    CREATE TABLE IF NOT EXISTS Products (
        product_id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        category TEXT NOT NULL,
        price REAL NOT NULL,
        stock INTEGER NOT NULL,
        supplier_id INTEGER NOT NULL,
        FOREIGN KEY (supplier_id) REFERENCES Suppliers(supplier_id)
    );
    """,

    "Orders": """
    CREATE TABLE IF NOT EXISTS Orders (
        order_id INTEGER PRIMARY KEY AUTOINCREMENT,
        customer_id INTEGER NOT NULL,
        employee_id INTEGER NOT NULL,
        order_date DATETIME NOT NULL,
        total_amount REAL NOT NULL,
        FOREIGN KEY (customer_id) REFERENCES Customers(customer_id),
        FOREIGN KEY (employee_id) REFERENCES Employees(employee_id)
    );
    """,

    "Order_Items": """
    CREATE TABLE IF NOT EXISTS Order_Items (
        order_item_id INTEGER PRIMARY KEY AUTOINCREMENT,
        order_id INTEGER NOT NULL,
        product_id INTEGER NOT NULL,
        quantity INTEGER NOT NULL,
        subtotal REAL NOT NULL,
        FOREIGN KEY (order_id) REFERENCES Orders(order_id),
        FOREIGN KEY (product_id) REFERENCES Products(product_id)
    );
    """,

    "Employees": """
    CREATE TABLE IF NOT EXISTS Employees (
        employee_id INTEGER PRIMARY KEY AUTOINCREMENT,
        first_name TEXT NOT NULL,
        last_name TEXT NOT NULL,
        email TEXT UNIQUE NOT NULL,
        phone_number TEXT NOT NULL,
        role TEXT NOT NULL,
        salary REAL NOT NULL
    );
    """,

    "Suppliers": """
    CREATE TABLE IF NOT EXISTS Suppliers (
        supplier_id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        contact_name TEXT NOT NULL,
        phone_number TEXT NOT NULL,
        email TEXT UNIQUE NOT NULL,
        address TEXT NOT NULL
    );
    """
}

# Execute table creation queries
for table, query in table_queries.items():
    cur.execute(query)

# Commit changes
con.commit()
print("All tables created successfully!")
```

```python
# Reconnect to the database
con = sqlite3.connect(db_path)
cur = con.cursor()

# Get all table names
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cur.fetchall()

# Print schema details for each table
print("\nDatabase Schema:")
for table in tables:
    table_name = table[0]
    print(f"\nüü¢ Table: {table_name}")

    # Get column details
    cur.execute(f"PRAGMA table_info({table_name});")
    columns = cur.fetchall()

    # Print column details
    print("{:<5} {:<20} {:<15} {}".format("ID", "Column Name", "Data Type", "Constraints"))
    print("-" * 50)

    for col in columns:
        col_id, col_name, col_type, not_null, default_val, pk = col
        constraints = []
        if pk:
            constraints.append("PRIMARY KEY")
        if not_null:
            constraints.append("NOT NULL")
        print(f"{col_id:<5} {col_name:<20} {col_type:<15} {' '.join(constraints)}")

# Close connection
con.close()
```
